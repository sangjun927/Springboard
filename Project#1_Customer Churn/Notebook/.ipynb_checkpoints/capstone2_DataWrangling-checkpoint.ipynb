## Overview
In this notebook, I will read 6 excel files and merge it by column 'Customer ID'. After succesfully merge the dataset, I am going to find missing values, review distribution, and check datatype to see whether dataset is valid enough or not.

## 1. Imports

```python
import pandas as pd
import matplotlib.pyplot as plt

data_churn = pd.read_excel("telco_data/Telco_customer_churn.xlsx")
data_status = pd.read_excel("telco_data/Telco_customer_churn_status.xlsx")
data_services = pd.read_excel("telco_data/Telco_customer_churn_services.xlsx")
data_location = pd.read_excel("telco_data/Telco_customer_churn_location.xlsx")
data_demographics = pd.read_excel("telco_data/Telco_customer_churn_demographics.xlsx")
data_populations = pd.read_excel("telco_data/Telco_customer_churn_population.xlsx")
```

## 2. Check the size of excel files

```python
print(data_churn.shape)
print(data_status.shape)
print(data_services.shape)
print(data_location.shape)
print(data_demographics.shape)
print(data_populations.shape)
```

## 3. Merge dataset

### 3-1. Rename customer id of data_churn

```python
# Rename customer id of data_churn. (It has a whitesapce.)
data_churn=data_churn.rename(columns={"CustomerID":"Customer ID"})
print(data_churn.info())
```

### 3-2. Merge 5 dataset by customer id

```python
data = pd.merge(pd.merge(pd.merge(pd.merge(data_churn, data_status, how='inner', on='Customer ID'), data_services, how='inner', on='Customer ID'), 
                data_location, how='inner', on='Customer ID'),data_demographics, how='inner', on='Customer ID')
```

### 3-3. Check the information of data

```python
print(data.info())
```

### 3-4. Drop duplicated columns

```python
# Drop duplicated columns
data.drop(data.filter(regex='_y$').columns, axis=1, inplace=True)
print(data.info())
```

### 3-5. Take out _x from column name

```python
# Take out _x from column name
data.columns = data.columns.str.replace(r'_x$', '')
print(data.info())
```

### 3-6. Merge population excel by Zip Code

```python
# merge population excel file to data
data = pd.merge(data, data_populations, how='inner', on='Zip Code')
data.info()
```

Finalized merged dataset has 61 columns composed of 8 float, 17 int, and 36 object.

## 4. Find missing value


```python
# Count the missing values
missing = pd.concat([data.isnull().sum(), 100 * data.isnull().mean()], axis=1)
missing.columns=['count', '%']
missing.sort_values(by='count', ascending = False)
```

It turns out 'Churn Reason' and 'Churn Category' have over 5000 missing values.

## 5. Review Distribution

```python
# Review Distribution
data.hist(figsize=(15,10))
plt.subplots_adjust(hspace=0.5);
```

So far, I couldn't find any skewed column.

## 7. Convert Datatype

### 7-1. Count values of 'Total Charges'

```python
data['Total Charges'].value_counts()
```

When I run the value counts of 'Total Charges', it gives me 11 " " values. These values has to be 0 before we convert to float.

### 7-2. Replace white space to 0

```python
data['Total Charges'] = data['Total Charges'].replace(" ", 0)
data['Total Charges'].value_counts()
```

Empty spaces have been removed and replaced to 0.00.

### 7.3 Convert datatype

```python
data['Total Charges'] = data['Total Charges'].astype(float)
print(data.info())
```

Total Charges has been converted to float.

## 8. Check Outliers

### 8-1. Create function to check min and max value of continuous variables.

```python
import numpy as np
# filter float and num
data_float = data.select_dtypes(include=[np.float])
data_int = data.select_dtypes(include=[np.int])

# find min max for all columns
def minMax(x):
    return pd.Series(index=['min','max'],data=[x.min(),x.max()])
```

```python
# min and max of float
data_float.apply(minMax)
```

```python
# min and max of int
data_int.apply(minMax)
```

Min value of Population seems suspicious. Let's check the observation which has Population 11.
